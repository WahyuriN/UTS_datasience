# Import Library

```python
import pandas as pd
import numpy as np
```

# Muat Dataset Titanic

```python

url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)
```

# Tampilkan head() dan info()

```python
print(df.head())
```

```python
print(df.info())
```

# Tampilkan jumlah baris dan kolom, dan tampilkan nama kolom

```python
print(df.shape, df.columns.tolist())
```

# Tampilkan missing values

```python
print(df.isnull().sum())
```

# Tampilkan persentase penumpang selamat

```python
print(df['Survived'].mean())
```

# Filter penumpang wanita kelas 1, jumlah yang selamat, dan Rata-rata usia

```python
wk1 = df[(df['Sex']=='female') &
(df['Pclass']==1)]
print(wk1['Survived'].mean(), wk1['Age'].mean())

```

# Menjalankan groupby('Pclass')['Survived'].mean()

```python
print(df.groupby('Pclass')['Survived'].mean())
```
